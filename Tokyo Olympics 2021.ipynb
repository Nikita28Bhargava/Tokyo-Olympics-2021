# Q.1 Which game was most played by the participants?

```python
plt.figure(figsize=(20,3))
sns.barplot(x=teams.Discipline.value_counts().index,y=teams.Discipline.value_counts().values)
plt.xlabel("DISCIPLINE")
plt.ylabel("FREQUENCY")
plt.xticks(rotation=45)
plt.show()
```

Ans. From the above graph we can see that maximum participants are from swimming sports.

#  Q.2 Which country has maximum number of participants?

```python
plt.figure(figsize=(20,3))
sns.barplot(x=teams.NOC.value_counts().index,y=teams.NOC.value_counts().values)
plt.xlabel("COUNTRIES")
plt.ylabel("FREQUENCY")
plt.xticks(rotation=90)
plt.show()
```

Ans. Maximum participants are from Japan. And the reason for this could be - Japan was hosting olymics this year.

# Q.3  What are total number of each gold, silver and bronze medals awarded? Are they equal?

```python
total_gold = medals["Gold"].sum()
total_silver = medals["Silver"].sum()
total_bronze = medals["Bronze"].sum()
print('Total Gold Medals',total_gold)
print('Total Silver Medals',total_silver)
print('Total Bronze Medals',total_bronze)
```

Ans.  The bronze medals are quite higher than the gold and silver medals. I believe it is because some of the events do not have a bronze medal match such as boxing.
